#### 1. Factorial

```python
test_case = [
    {
        "Input" : 1,
        "Output" : 1, 
    },
    {
        "Input" : 5,
        "Output" : 120, 
    }
]

class Solution():

    def __init__(self):
        pass

    def Factorial(self, nums):
        """
        write code here
        function factorial get n input and return the result of factorial
        """



        return nums
    


runner = Solution()
total = len(test_case)
correct = 0
for case in test_case:
    rst = runner.Factorial(case['Input'])
    if rst != case["Output"]:
        print(f"Error in case Input : {case['Input']} Output : {case['Output']}")
    else: 
        correct += 1

print(f"\nscore : {correct}/{total}")
```
