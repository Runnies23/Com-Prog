```python
def round_taxi(nums):
    if nums % 5 == 0:
        return nums
    if nums % 10 < 5:
        return ((nums // 10) * 10) + 5
    elif nums % 10 > 5:
        return ((nums // 10) + 1) * 10
```

```python
round_taxi(50)
```

```python
total = 0
for i in data:
    total += round_taxi(i)
```
