```python
print(data)
sum_li = []
for i in data:
    new_li = i.split(" ")
    # print(new_li)
    first = int(new_li[0])
    seco = int(new_li[2])
    opp = new_li[1]

    if opp == "+":
        res = first + seco
    elif opp == "-":
        res = first - seco
    elif opp == "*":
        res = first * seco
    elif opp == "/":
        res = first / seco
    sum_li.append(res) 
print(sum(sum_li))
```
